# Ollama

Ollama supports embedding models, making it possible to build retrieval augumented generation(RAG) applications that combine text prompts with existing documents or other data.

```python
# !pip install langchain_community
```

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="gemma2:2b") #by default it uses llama2
)
```

```python
embeddings
```

```python
rl=embeddings.embed_documents(
    [
        "Alpha is the first letter of Greek alphabet",
        "Beta is the second letter of Greek alphabet",
    ]
)
```

```python
rl[0]
```

```python
len(rl[0])  
```

```python
embeddings.embed_query("What is the second letter of Greek alphabet")
```
